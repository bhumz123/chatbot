```python
wnlen = nltk.stem.WordNetLemmatizer()
```

```python
def perform_lemmatizer(tokens):
    return[wnlen.lemmatize(token) for token in tokens ]
```

```python
pr = dict((ord(punctuation), None)for punctuation in string.punctuation)
```

```python
#tokenizer to be used
def get_processed_text(document):
    return perform_lemmatizer(nltk.word_tokenize(document.lower().translate(pr)))
```

```python
greeting_inputs=("hey","hello","good morning","good evening","morning")
greeting_reponses= ("hey", "how are you","hello","hey, welcome dear")
def generate_greeting_response(greeting):
    for token in greeting.split():
        if token.lower() in greeting_inputs:
            return random.choice(greeting_responses)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
continue_flag = True
print("hello I am a Data Science Bot , you may ask your query.")
while(continue_flag == True):
    human = input()
    human = human.lower()
    if human!= "bye":
        if human == "thanks" or human == "thank you":
            continue_flag = False
            print("Most Welcome")
        else:
            if generate_greeting_response(human) != None:
                print(" DS Bot "+ generate_greeting_response(human))
            else:
                print("DS Bot" , end ="")
                print(generate_response(human))
                sen.remove(human)
    else:
            continue_flag = False
            print("Good Bye")
```
